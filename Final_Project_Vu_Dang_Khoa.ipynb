```python
#Demographic Analysis
df.age.value_counts()
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px


# Tính số lượng khách hàng theo độ tuổi
age_counts = df['age'].value_counts().sort_index()

# Vẽ biểu đồ cột
plt.figure(figsize=(30, 6))
bars = plt.bar(age_counts.index, age_counts, color='orange')

for bar, age in zip(bars, age_counts.index):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', color='black', fontweight='bold')
    plt.text(bar.get_x() + bar.get_width()/2, -7, str(age), ha='center', va='bottom', color='black')

plt.title('Age Analysis', fontweight='bold')
plt.xlabel('Age', fontweight='bold')
plt.ylabel('Quantity', fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.7)
plt.gca().set_facecolor('#F5F5F5')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

gender_counts = df['Gender'].value_counts()

# Vẽ biểu đồ tròn
plt.figure(figsize=(10, 9))
plt.title('Gender', fontweight='bold')
_, _, autotexts = plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['darkorange','orange','gold','moccasin'], startangle=90)

# Chèn giá trị số lượng vào từng phần của biểu đồ
for i, (autotext, count) in enumerate(zip(autotexts, gender_counts)):
    autotext.set_text(f'{count}\n{gender_counts[i] / gender_counts.sum() * 100:.1f}%')

plt.show()
```

```python
#Categories
from wordcloud import WordCloud
import random
from PIL import Image
import matplotlib
import seaborn as sns

# Tách thành nhiều dòng

# Tạo hàm tùy chỉnh để tách giá trị sau dấu ';'
def split_values(value):
    return [v.strip() for v in value.split(';')]

# Chuyển đổi giá trị trong cột 'Purchase_Categories' thành chuỗi
df['Purchase_Categories'] = df['Purchase_Categories'].astype(str)

# Tách giá trị sau dấu ';' thành dòng mới
df['Purchase_Categories'] = df['Purchase_Categories'].str.split(';')

# Mở rộng DataFrame để có một hàng cho mỗi giá trị
df_expanded = df.explode('Purchase_Categories')

# Loại bỏ dấu '' khỏi các giá trị
df_expanded['Purchase_Categories'] = df_expanded['Purchase_Categories'].apply(lambda x: x.strip()).replace('', pd.NA)

# Loại bỏ các giá trị rỗng
df_expanded = df_expanded.dropna(subset=['Purchase_Categories'])

# Tạo đoạn văn bản cho WordCloud
text = ' '.join(df_expanded['Purchase_Categories'].astype(str))

wordcloud = WordCloud(background_color = 'white', width = 500,  height = 200,colormap='gist_heat', max_words = 150).generate(text)

plt.figure(figsize=(5,5))
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```

```python
def create_count_plot(data, x_col, hue_col, title, category_mapping=None, group_threshold=None):
    # Define the neon_palette
    neon_palette = ['#485769', '#232f3e', '#f3a847', "orange", "#3d144c"]

    # Apply category mapping and grouping if provided
    if category_mapping:
        data[x_col] = data[x_col].map(category_mapping)
    if group_threshold:
        category_counts = data[x_col].value_counts()
        less_frequent_categories = category_counts[category_counts < group_threshold].index
        data.loc[data[x_col].isin(less_frequent_categories), x_col] = 'Other Categories'

    # Create the count plot
    plt.figure(figsize=(10, 6))
    ax = sns.countplot(data=data, x=x_col, hue=hue_col, palette=neon_palette)
    ax.set_facecolor('#F6F5F4')
    ax.set_title(title, fontsize=16, pad=20)
    ax.set_xlabel(x_col, fontsize=12)
    ax.set_ylabel("Count", fontsize=12)
    ax.tick_params(axis='x', labelrotation=0)
    ax.grid(True, linestyle='--', alpha=0.5)
    sns.despine(right=True)
    sns.despine(offset=5, trim=True)

    # Adding percentage annotations
    total = float(len(data))
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x() + p.get_width() / 2., height + 3, '{:1.1f}%'.format((height / total) * 100),
                ha="center", fontsize=10, bbox=dict(facecolor='none', edgecolor='black', boxstyle='round', linewidth=0.5))

    plt.tight_layout()
    plt.show()

# Define your category mapping and group threshold
category_mapping = {
    'Beauty and Personal Care': 'Beauty and Personal Care',
    'Clothing and Fashion': 'Clothing and Fashion',
    'Groceries and Gourmet Food': 'Groceries and Gourmet Food',
    'Home and Kitchen': 'Home and Kitchen'
    # Add more mappings as needed
}
group_threshold = 10
```

```python

create_count_plot(df, x_col='Purchase_Categories', hue_col='Purchase_Frequency', title='Purchase Frequency by Categories',
                  category_mapping=category_mapping, group_threshold=group_threshold)
```

# Feedback Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/Phan tich feedback cua Amazon.csv')

df_sorted = df.sort_values(by='Quantity', ascending=False)

top_4_values = df_sorted.head(4)

other_values = pd.DataFrame({
    'Improvement Areas': ['Others'],
    'Quantity': [df_sorted['Quantity'][4:].sum()]
})

combined_df = pd.concat([top_4_values, other_values])
plt.figure(figsize=(10, 9))
plt.title('Feedback Analysis', fontweight='bold')

explode = (0, 0, 0, 0, 0) if len(df_sorted) > 4 else (0, 0, 0, 0, 0)
_, _, autotexts = plt.pie(combined_df['Quantity'], labels=combined_df['Improvement Areas'], autopct='%1.1f%%', colors=['darkorange', 'orange', 'gold', 'moccasin', 'coral'], startangle=90, explode=explode)

for autotext in autotexts:
    autotext.set_color('black')
plt.show()

```
